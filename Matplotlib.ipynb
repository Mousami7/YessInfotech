```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.xlabel("days")
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x, y, color = "m", linewidth= 5, linestyle = "dashdot")
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 35, 20]
min_t = [43, 45, 40, 39, 30, 50, 38]
avg_t = [45, 48, 50, 25, 35, 46, 48]
```

```python
plt.plot(days, max_t, label = "max")
plt.plot(days, min_t, label = "min")
plt.plot(days, avg_t, label = "average")

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.legend(loc = "best")
plt.title("Weather")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(days, max_t, label = "max")
plt.plot(days, min_t, label = "min")
plt.plot(days, avg_t, label = "average")

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.legend(loc = "upper right", shadow = True, fontsize = 'large')
plt.title("Weather")
plt.grid()
plt.show()
```

# Bar Plot

```python
company = ["Google", "Amazon", "Microsoft", "Meta"]
revenue = [90, 136, 89, 27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos, revenue, label = "Revenue")
plt.xticks(xpos, company)
plt.ylabel("Revenue(Bln)")
plt.title("US Technology Stocks")
plt.legend()
plt.show()
```

# Multiple Bars showing revenue and profit of major US tech Companies

```python
profit = [40, 2, 34, 12]
```

```python
plt.bar(xpos-0.2, revenue, width = 0.4, label = "Revenue")
plt.bar(xpos+0.2, profit, width = 0.4, label = "Profit")

plt.xticks(xpos, company)
plt.ylabel("Revenue (Bln)")
plt.title("US Technology Stock")
plt.legend()
plt.show()
```

# Horizontal bar chart using barh function

```python
plt.barh(xpos, revenue, label = "Revenue")

plt.yticks(xpos, company)
plt.ylabel("Revenue(Bln)")
plt.title("US Technology Stock")
plt.legend()
plt.show()
```

### We have sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre- diabetic and diabetic

```python
blood_sugar = [113, 150, 90, 88, 89, 100, 105, 110, 77, 129, 130]
plt.hist(blood_sugar, rwidth = 0.8)
plt.show()
```

```python
plt.hist(blood_sugar, rwidth = 0.8, bins = 4)
plt.show()
```

```python
plt.hist(blood_sugar, rwidth = 0.8, bins = [80, 90, 100, 110, 120, 130, 140, 150])
plt.show()
```

## Histogram showing normal, pre-diabetic and diabetic patients distribution

1. 80 - 100: Normal
2. 100 -125: Pre-diabetic
3. 125 - 150: diabetic

```python
plt.xlabel("Sugar level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins= [80, 100, 125, 150], rwidth= 0.5, color = 'crimson', alpha = 0.6)
plt.show()
```

# Multiple data samples in a histogram

```python
plt.xlabel("Sugar level")
plt.ylabel("NUmber of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 99, 88, 93, 135, 77, 82, 129]
blood_sugar_women = [67, 90, 89, 120, 75, 120, 112, 100, 122, 135, 140, 125]

plt.hist([blood_sugar_men, blood_sugar_women], bins = [80, 100, 125, 150],  
         rwidth= 0.95, color= ["lightblue", "royalblue"], label=['men', 'women'])
plt.legend()
plt.show()
```
